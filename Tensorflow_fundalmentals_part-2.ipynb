```python
# 021 changing the datatype of tensors
import tensorflow as tf 
a=tf.constant([3,3])
a.dtype
```

```python
c=tf.constant([3.3,5.3])
c.dtype

# change from float32 to   float16
b=tf.cast(a,dtype=tf.float16) #same as we were using in other language
b.dtype

e=tf.cast(c,dtype=tf.int32)
e
```

```python
# 022 Tensor Aggregation : the collecting of units;
# aggregating tensors = condensing them from multiple values down to a smaller amount of values
# get the absoulte values
import tensorflow as tf
d=tf.constant([-2,3,3])
tf.abs(d)  # convert the negative to positive
```

```python
# lets go through the following forms of aggregation
# get the minimum , get the maximum , get the mean of tensor
import numpy as np
sri=tf.constant(np.random.randint(0,100,50))
sri
```

```python
tf.size(sri),sri.shape, sri.ndim
```

```python
tf.reduce_min(sri)
```

```python
tf.reduce_max(sri)
```

```python
tf.reduce_mean(sri)
```

```python
tf.reduce_sum(sri)
```

```python
# 023 #find the variance of our tensor
import tensorflow_probability as tfp
tfp.stats.variance(sri)   # another way of finding the variance
tf.math.reduce_variance(sri)  
```

```python
 # find the standard deviation
sri=tf.cast(sri,dtype=tf.float32)  # without this will get error because it must be complex or real numbers
tf.math.reduce_std(sri)
```

```python
# 024 Find the positional maximum and minimum
# at which row or column we need to find the max or min
f=tf.random.uniform(shape=[50])
f
```

```python
# find the positional maximum
tf.argmax(f)
```

```python
#index on our largest value position
f[tf.argmax(f)]
```

```python
assert f[tf.argmax(f)]==tf.reduce_max(f)
```

```python
tf.argmin(f)  # gives the minimum value position
```

```python
# 025 Squeezing a tensor (removing all single dimensions)
#create a tensor to get started
import tensorflow as tf
tf.random.set_seed(34)
g=tf.constant(tf.random.uniform(shape=[50]),shape=(1,1,1,1,50))
g
```

```python
g.shape
```

```python
# 026 One hot encoding tensors - Is form of numerical encoding.(for more go to machine learning mastery website)
# create a list of indices
a=[4,3,2,1,0]
# one hot encode our list of indices
tf.one_hot(a,depth=4)  # if depth is less than the size of indices then it will not consider that
```

```python
# speicfy the custom values for one hot encoding
b=[2,1,0]
tf.one_hot(b,depth=3,on_value="blue is color",off_value="its green")
```

```python
# 027 trying out more tensor math operations
#squraing , log, square root
sri=tf.range(1,10)
sri
```

```python
sq_sri=tf.square(sri)
sqr_sri=tf.math.sqrt(tf.cast(sri,dtype=tf.float32))
log_sri=tf.math.log(tf.cast(sri,dtype=tf.float32))
sq_sri, sqr_sri, log_sri
```
