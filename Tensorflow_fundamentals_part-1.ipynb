```python
#shuffle the order of elements in a tensor(valuabe for when you want to shuffle your data so the inherent order doesn't effect learning)
not_shuffle=tf.constant([[12,4,2],
                        [1,3,2],
                        [1,4,2]])
not_shuffle.ndim
```

```python
tf.random.shuffle(not_shuffle)
tf.random.shuffle(not_shuffle,seed=33)  # it just operational-level seed
```

```python
tf.random.set_seed(43)  # global level seed
tf.random.shuffle(not_shuffle,seed=33)  # it looks like if we want our shuffled tensors to be in the same order
#,we've got to use the global level random seed as well as operation level random seed:
# rule 4 if both the global and the operation seed are set both seeds are used in conjuction to deteremine the ranoom sequenece;
```

```python
#getting information from tensors
#attributes:shape,rank,axis or dimension,size
#creating the rank 4 tensor(4 dimension)
rank4=tf.zeros(shape=[2,3,4,6])
rank4[0]
rank4.shape , rank4.ndim , tf.size(rank4)
#rank4
```

```python
#get various attributes of our tensors;
print("datatppe of everyelement",rank4.dtype)
print(rank4.shape)
print(rank4.ndim)
print(rank4.shape[0])
print(rank4.shape[-1])
print(tf.size(rank4).numpy())
```

```python
# 016 indexing and expanding tensors
# eg ten=[1,2,3,4]
#    ten[:2]
# get the first 2 elements of each dimension
rank4[:2,:2,:2,:2 ]
```

```python
# get the first element from each dimension from each index except for the final one
rank4[:1,:1,:1,:]
```

```python
#create a rank 2 tensor(2 dimension)
rank2=tf.constant([[1,2],[3,4]])
rank2
```

```python
#get the last item of each of our rank 2 tensor
rank2.shape
rank2[:,-1]    # rows*columns
```

```python
#add in extra dimension to our rank 2 tensor
rank3=rank2[...,tf.newaxis]  # instead of :,: -- ...
rank3
```

```python
#alternative 
tf.expand_dims(rank2,axis=0)   # 0 means expand the first axis
tf.expand_dims(rank2,axis=-1)  # -1 means expand the final axis
```
